```python
from neck_correctify import load_img_and_ann
import matplotlib.pyplot as plt 
import math

from geometry_tools import line_segments_offset
```

```python
keypoint_list = ["left_shoulder", "right_shoulder", "left_mouse", "right_mouse", "nose", "left_eye",
                 "right_eye", "left_ear", "right_ear","chi"]
```

```python
folder = "./测试素材/测试人物/"
ann_folder = load_img_and_ann(folder)
```

```python
# 计算中点
def middle_point(p1,p2):
    return ( (p1[0] + p2[0]) /2 ,(p1[1] + p2[1]) /2   )
```

```python
markersize = 0.8
for idx , item in enumerate( ann_folder.items()):
    img_ = item[1][0]
    ann_ = item[1][1]
    plt.figure()
    plt.imshow(img_)
    for kpn in ann_.keys():
        kps_ = ann_[kpn]
        for kp_ in kps_:
            plt.plot( math.floor(kp_[0]) , math.floor(kp_[1]), 'ro' ,markersize = markersize)
    
    
        # 肩膀中点
    mid_p = middle_point( ann_['left_shoulder'][0], ann_['right_shoulder'][0]  )
    plt.plot( math.floor(mid_p[0]) , math.floor(mid_p[1]), 'go' ,markersize =markersize)
    plt.savefig('./temp/' +item[0] +'.jpg',dpi=600  )
    plt.show()
    
    break 
```

```python
#  脸颊点偏移
markersize = 0.1
img_4_2 = ann_folder['4 (2)'][0]
chi_points = ann_folder['4 (2)'][1]['chi']
chi_points = sorted(chi_points ,key= lambda x : x[0])
chi_offset = line_segments_offset(chi_points , 60 )

plt.figure()
plt.imshow(img_4_2)
for kp_ in chi_points:
    plt.plot( math.floor(kp_[0]) , math.floor(kp_[1]), 'ro' ,markersize = markersize)
for kp_ in chi_offset:
    plt.plot( math.floor(kp_[0]) , math.floor(kp_[1]), 'go' ,markersize = markersize)
plt.savefig('./temp/'  +' chi_offset.jpg',dpi=600  )
plt.show()
```
