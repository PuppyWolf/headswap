```python
# 估算头部截取
# 肩部宽度作为 宽， 长为从 脸颊最下面点 到鼻子 长度的  两倍。

def get_head_mask(ann):
    left_shoulder = ann['left_shoulder'][0]
    right_shoulder = ann['right_shoulder'][0]
    nose = ann['nose'][0]
    chi = ann['chi']
    chi_points = sorted(chi ,key= lambda x : x[0])
    
    lowest_chi = chi_points[5] # 下巴点
    mid_p = middle_point( left_shoulder, right_shoulder  )
        
    
    width =  abs( left_shoulder[0] -  right_shoulder[0] )
    height = 2* abs( nose[1] - mid_p[1]  )
    
    delta = 5
    left_upper  = [ mid_p[0] - width/2,   max(lowest_chi[1] + delta - height , 0) ] 
    right_bottom = [ left_upper[0] + width , lowest_chi[1] + delta ] 
    
    return [ math.floor(left_upper[0]) , math.floor(left_upper[1] ), math.floor(right_bottom[0] ), math.floor(right_bottom[1])]
```

```python
# 显示截取效果
ann_4_2 = ann_folder['4 (2)'][1]
head_bbox = get_head_mask(ann_4_2)

img_4_2 = ann_folder['4 (2)'][0]
head_ = img_4_2.crop(head_bbox)
#mask = 255*np.ones( shape = ( head_bbox[2]- head_bbox[0]  , head_bbox[3] - head_bbox[1]),dtype = np.uint8) 
temp_4_2 = img_4_2.copy()
temp_4_2.paste( (255,255,255),tuple( head_bbox) )
plt.imshow(temp_4_2)
plt.show()
```

```python
# 计算 原图 在 目标图中的 最佳位置
Q = ann_4_3['chi'] #  目标中关键点的位置
P = ann_4_2['chi']
```

```python
def cost_func_affine(T):
    cost = 0.0
    TM =  np.array( [ [ T[0], T[1], T[4] ] , [ T[2] , T[3] ,T[5]  ] , [0, 0, 1]   ]  )
    P_t_all = []
    for i in range(11):
        P_t = np.matmul(TM , np.array( [ P[i][0],P[i][1] ,1 ]).T) # note : homogeneus coordinate
       # P_t_all.append(P_t)
        cost +=  (P_t[0] - Q[i][0]) ** 2  + ( P_t[1] - Q[i][1]) ** 2
        #print(cost)
    
 
    return cost
```

```python
source_image_mask = np.zeros(shape = img_4_2.size+(3,) , dtype = np.uint8 )
source_image_mask = Image.fromarray(source_image_mask)
source_image_mask.paste(head_  ,box = head_bbox_2 )
```

```python
from scipy.optimize import minimize
T0 = np.array( [10,10,10,1,1,1])
res = minimize(cost_func_affine, T0, method='nelder-mead', options={'xatol': 1e-8, 'disp': True})
print(res.success)
print(res.x)
```

```python
T=  res.x
TM =  np.array( [ [ T[0], T[1], T[4] ] , [ T[2] , T[3] ,T[5]  ] , [0, 0, 1]   ]  )
for i in range(11):
     P_t = np.matmul(TM ,np.array( [ P[i][0],P[i][1] ,1 ]).T )
     print(P_t[:2])
```
