```python
from neck_correctify import load_img_and_ann
import matplotlib.pyplot as plt 
import math
```

```python
keypoint_list = ["left_shoulder", "right_shoulder", "left_mouse", "right_mouse", "nose", "left_eye",
                 "right_eye", "left_ear", "right_ear"]
```

```python
folder = "./测试素材/测试人物/"
ann_folder = load_img_and_ann(folder)
```

```python
# 计算中点
def middle_point(p1,p2):
    return ( (p1[0] + p2[0]) /2 ,(p1[1] + p2[1]) /2   )
```

```python
for idx , item in enumerate( ann_folder.items()):
    img_ = item[1][0]
    ann_ = item[1][1]
    plt.figure()
    plt.imshow(img_)
    for kpn in keypoint_list:
        kp_ = ann_[kpn]
        plt.plot( math.floor(kp_[0]) , math.floor(kp_[1]), 'ro' ,markersize = 2)
        
        # 肩膀中点
        mid_p = middle_point( ann_['left_shoulder'], ann_['right_shoulder']  )
        plt.plot( math.floor(mid_p[0]) , math.floor(mid_p[1]), 'go' ,markersize = 2)
    plt.savefig('./temp/' +item[0] +'.jpg',dpi=600  )
    plt.show()
```
